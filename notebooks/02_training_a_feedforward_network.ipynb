# Training a Feed Forward Network

This experiment extends the single neuron example into a multi-layer neural network.

The goal is to demonstrate how neural networks gain expressive power by composing
multiple nonlinear transformations.

The model architecture is:

$$
x \rightarrow Dense \rightarrow Activation \rightarrow Dense \rightarrow Activation \rightarrow Dense \rightarrow \hat{y}
$$

Unlike the single neuron experiment, this network contains hidden layers that
learn intermediate representations of the input.

The complete training cycle remains the same:

$$
\text{forward}
\rightarrow
\text{loss}
\rightarrow
\text{backward}
\rightarrow
\text{parameter update}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

from nnlab.kernels import LogisticKernel

from nnlab.activations import ParameterizedActivation

from nnlab.layers import (
    Dense,
    ActivationLayer,
)

from nnlab.models import FeedForward

from nnlab.losses import MeanSquaredError

from nnlab.optimizers import SGD
```

```python
x = np.linspace( -3.0, 3.0, 300, ).reshape( -1, 1, )

target = x**2
```

```python
plt.figure(figsize=(8,4))

plt.plot(
    x,
    target,
)

plt.title(
    "Target Function: y = x²"
)

plt.xlabel(
    "Input"
)

plt.ylabel(
    "Output"
)

plt.grid(True)

plt.show()
```

```python
activation_1 = ActivationLayer(
    activation=ParameterizedActivation(
        kernel=LogisticKernel(),
        kernel_scale=1.0,
    )
)

activation_2 = ActivationLayer(
    activation=ParameterizedActivation(
        kernel=LogisticKernel(),
        kernel_scale=1.0,
    )
)
```

## Network Architecture

The network contains:

- one input feature,
- two hidden layers,
- one output neuron.

The hidden layers transform the input into intermediate features.

The final layer combines those features into the predicted output.

```python
model = FeedForward(
    layers=[
        Dense(
            input_size=1,
            output_size=16,
        ),

        activation_1,

        Dense(
            input_size=16,
            output_size=16,
        ),

        activation_2,

        Dense(
            input_size=16,
            output_size=1,
        ),
    ],
)
```

```python
prediction = model.forward(
    x,
)

plt.figure(figsize=(8,4))

plt.plot(
    x,
    target,
    label="Target",
)

plt.plot(
    x,
    prediction,
    label="Initial Prediction",
)

plt.legend()

plt.grid(True)

plt.show()
```

```python
loss = MeanSquaredError()

optimizer = SGD(
    learning_rate=0.01,
)
```

```python
epochs = 2000

history = []

for epoch in range(
    epochs,
):

    prediction = model.forward(
        x,
    )

    error = loss.forward(
        prediction,
        target,
    )

    gradient = loss.derivative(
        prediction,
        target,
    )

    model.backward(
        gradient,
    )

    optimizer.step(
        model.parameters(),
        model.gradients(),
    )

    history.append(
        error,
    )
```

```python
plt.figure(figsize=(8,4))

plt.plot(
    history,
)

plt.title(
    "Training Loss",
)

plt.xlabel(
    "Epoch",
)

plt.ylabel(
    "MSE",
)

plt.grid(True)

plt.show()
```

```python
prediction = model.forward(
    x,
)

plt.figure(figsize=(8,4))

plt.plot(
    x,
    target,
    label="Target",
)

plt.plot(
    x,
    prediction,
    label="Prediction",
)

plt.legend()

plt.title(
    "Learned Function",
)

plt.xlabel(
    "Input",
)

plt.ylabel(
    "Output",
)

plt.grid(True)

plt.show()
```

```python
hidden = model.layers[0].forward(
    x,
)

plt.figure(figsize=(8,4))

plt.plot(
    x,
    hidden,
)

plt.title(
    "Hidden Layer Features",
)

plt.xlabel(
    "Input",
)

plt.ylabel(
    "Neuron Response",
)

plt.grid(True)

plt.show()
```

## Observations

The network successfully approximates a nonlinear function that cannot be
represented by a single neuron.

The additional layers allow the model to create intermediate representations
and combine them into a more expressive final prediction.
