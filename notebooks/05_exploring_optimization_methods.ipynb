# Exploring Optimization Methods

A neural network learns by adjusting its parameters to reduce a loss function.

The optimization algorithm determines how those parameter updates are performed.

Previous experiments changed:

- network architecture,
- activation functions,
- model capacity.

This experiment keeps the neural network fixed and investigates how different
optimization methods influence the learning process.

Every experiment uses:

- the same dataset,
- the same network architecture,
- the same activation function,
- the same loss function.

Only the optimizer changes.

```python
import numpy as np
import matplotlib.pyplot as plt

from nnlab.kernels import LogisticKernel

from nnlab.activations import ParameterizedActivation

from nnlab.layers import (
    Dense,
    ActivationLayer,
)

from nnlab.models import FeedForward

from nnlab.losses import MeanSquaredError

from nnlab.optimizers import SGD

from nnlab.training import Trainer
```

```python
x = np.linspace( -3.0, 3.0, 400, ).reshape( -1, 1, )

target = ( np.sin(2 * x) + 0.25 * x )
```

## Fixed Network Architecture

All optimizers will train the same model.

The only variable is the parameter update rule.

```python
def build_model():

    return FeedForward(
        layers=[
            Dense(
                input_size=1,
                output_size=16,
            ),

            ActivationLayer(
                ParameterizedActivation(
                    kernel=LogisticKernel(),
                )
            ),

            Dense(
                input_size=16,
                output_size=16,
            ),

            ActivationLayer(
                ParameterizedActivation(
                    kernel=LogisticKernel(),
                )
            ),

            Dense(
                input_size=16,
                output_size=1,
            ),
        ],
    )
```

```python
optimizers = {
    "SGD lr=0.001":
        SGD(
            learning_rate=0.001,
        ),

    "SGD lr=0.01":
        SGD(
            learning_rate=0.01,
        ),

    "SGD lr=0.1":
        SGD(
            learning_rate=0.1,
        ),
}
```

```python
def train_with_optimizer(
    optimizer,
    epochs=5000,
):

    model = build_model()

    trainer = Trainer(
        model=model,
        loss=MeanSquaredError(),
        optimizer=optimizer,
    )

    history = trainer.fit(
        x,
        target,
        epochs=epochs,
    )

    return model, history
```

```python
results = {}

for name, optimizer in optimizers.items():

    model, history = train_with_optimizer(
        optimizer,
    )

    results[name] = {
        "model": model,
        "history": history,
    }
```

```python
plt.figure(figsize=(8,4))

for name, result in results.items():

    plt.plot(
        result["history"],
        label=name,
    )

plt.title(
    "Optimizer Convergence"
)

plt.xlabel(
    "Epoch"
)

plt.ylabel(
    "MSE"
)

plt.yscale(
    "log",
)

plt.legend()

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(
    x,
    target,
    "k--",
    linewidth=3,
    label="Target",
)

for name, result in results.items():

    prediction = result["model"].forward(x)

    plt.plot(
        x,
        prediction,
        label=name,
    )

plt.title(
    "Final Predictions by Optimizer"
)

plt.xlabel(
    "Input"
)

plt.ylabel(
    "Output"
)

plt.legend()

plt.grid(True)

plt.show()
```

## Observations

Optimization determines how efficiently a neural network learns.

A model may have sufficient capacity to represent a function, but the optimizer
controls how effectively the parameters move toward a useful solution.

Important factors include:

- learning rate,
- update strategy,
- gradient scaling,
- convergence speed.

There is no universally best optimizer. The appropriate choice depends on the
model, dataset, and learning objective.

The following experiments move from controlled mathematical studies into
practical applications, where these design choices must work together.
